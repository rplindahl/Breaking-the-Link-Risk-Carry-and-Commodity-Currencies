# Create commodity price index

```python
import pandas as pd
import numpy as np

# --- 1. Data Loading ---
print("--- Step 1: Loading Data ---")
try:
    df_exports = pd.read_csv("Trade/final_commodity_export_summary_all_years_WITH_EMU.csv")
    df_prices = pd.read_excel("Commodity_prices/all_commodity_prices_merged.xlsx", index_col="Date")
    print("Successfully loaded export weights and merged prices.")
except FileNotFoundError as e:
    print(f"Error loading files: {e}")
    exit()

# --- 2. Data Preparation ---
print("\n--- Step 2: Preparing Data ---")
weight_cols = ['Year', 'Country name'] + [col for col in df_exports.columns if 'Share_' in col and '_percent_of_commodity' in col]
df_weights = df_exports[weight_cols].copy()

cleaned_weight_cols = {}
for col in df_weights.columns:
    if 'Share_' in col:
        commodity_name = col.replace('Share_', '').replace('_percent_of_commodity', '')
        cleaned_weight_cols[col] = commodity_name
df_weights.rename(columns=cleaned_weight_cols, inplace=True)
print("Cleaned and standardized weight column names.")

df_prices.index = pd.to_datetime(df_prices.index)

# --- 3. Price Normalization ---
print("\n--- Step 3: Normalizing Prices ---")
base_date = '2016-01-05'
try:
    base_prices = df_prices.loc[base_date]
    normalized_prices = (df_prices / base_prices) * 100
    print(f"Prices normalized to base date: {base_date}")
except KeyError:
    print(f"FATAL ERROR: Base date '{base_date}' not found.")
    exit()

# --- 4. Define Special Mappings ---
print("\n--- Step 4: Defining Special Price Mappings ---")
europe_countries = ['Austria', 'Belgium', 'Cyprus', 'Finland', 'France', 'Germany', 'Greece', 'Ireland', 'Italy', 'Luxembourg', 'Malta', 'Netherlands', 'Portugal', 'Slovakia', 'Slovenia', 'Spain', 'Estonia', 'Latvia', 'Lithuania', 'EMU']
north_america_countries = ['USA', 'Canada', 'Mexico']
price_mapping = {
    'coal': {'default': 'Price_Coal_coal_price_1', 'South Africa': 'Price_Coal_coal_price_2', 'region_europe': 'Price_Coal_coal_price_3'},
    'corn': {'default': 'Price_Corn_corn_price_1', 'region_europe': 'Price_Corn_corn_price_2'},
    'crude_oil': {'default': 'Price_Crude_Oil_crude_oil_price_2', 'region_north_america': 'Price_Crude_Oil_crude_oil_price_1'},
    'natural_gas': {'default': 'Price_Natural_Gas_natural_gas_price_3', 'United Kingdom': 'Price_Natural_Gas_natural_gas_price_2', 'region_europe': 'Price_Natural_Gas_natural_gas_price_1'},
    'palm_oil': {'default': 'Price_Palm_Oil_palm_oil_price_2', 'Indonesia': 'Price_Palm_Oil_palm_oil_price_1'}
}

# --- 5. Main Index Calculation Loop ---
print("\n--- Step 5: Calculating Commodity Price Indices ---")
all_countries = df_weights['Country name'].unique()
final_indices = {}

available_prices = {col.lower().replace('_', '').replace('price', ''): col for col in normalized_prices.columns}

for country in all_countries:
    print(f"Processing index for: {country}")
    country_weights = df_weights[df_weights['Country name'] == country].set_index('Year')
    country_daily_index_parts = []

    for year in country_weights.index:
        start_date, end_date = f'{year}-01-01', f'{year}-12-31'
        prices_this_year = normalized_prices.loc[start_date:end_date]
        if prices_this_year.empty: continue
        
        weights_this_year = country_weights.loc[year]
        
        # --- THIS IS THE MODIFIED LOGIC ---
        
        # This will hold the weighted price series for this year
        weighted_series_for_year = pd.DataFrame(index=prices_this_year.index)
        # This will hold the sum of available weights FOR EACH DAY
        sum_of_available_weights_daily = pd.Series(0.0, index=prices_this_year.index)

        for commodity, weight in weights_this_year.items():
            if commodity in ['Country name'] or weight == 0 or pd.isna(weight):
                continue

            price_col_to_use = None
            
            if commodity in price_mapping:
                # ... (special mapping logic remains the same) ...
                if country in price_mapping[commodity]: price_col_to_use = price_mapping[commodity][country]
                elif 'region_europe' in price_mapping[commodity] and country in europe_countries: price_col_to_use = price_mapping[commodity]['region_europe']
                elif 'region_north_america' in price_mapping[commodity] and country in north_america_countries: price_col_to_use = price_mapping[commodity]['region_north_america']
                else: price_col_to_use = price_mapping[commodity]['default']
            else:
                clean_commodity_name = commodity.lower().replace('_', '').replace(' ', '')
                try:
                    matching_key = next(key for key in available_prices.keys() if clean_commodity_name in key)
                    price_col_to_use = available_prices[matching_key]
                except StopIteration:
                    continue
            
            if price_col_to_use in prices_this_year.columns:
                # Get the price series for this commodity
                price_series = prices_this_year[price_col_to_use]
                
                # Calculate the weighted series for this commodity
                weighted_series_for_year[commodity] = price_series * (weight / 100)
                
                # For each day this price is available, add its weight to our daily sum
                # .notna() creates a boolean series (True where price exists)
                # We multiply by the weight to add the weight only on those days
                sum_of_available_weights_daily += price_series.notna() * (weight / 100)

        # Sum the weighted price series across all commodities for each day
        numerator = weighted_series_for_year.sum(axis=1)
        
        # The denominator is now our dynamic daily sum of available weights
        denominator = sum_of_available_weights_daily
        
        # Avoid division by zero on days where no prices are available at all
        denominator.replace(0, np.nan, inplace=True)
        
        # Calculate the final, correctly normalized index for the year
        yearly_index = numerator / denominator
        # --- END OF MODIFIED LOGIC ---
        
        country_daily_index_parts.append(yearly_index)

    if country_daily_index_parts:
        final_indices[country] = pd.concat(country_daily_index_parts).dropna()

# --- 6. Final Assembly and Saving ---
print("\n--- Step 6: Assembling Final DataFrame and Saving ---")
final_index_df = pd.DataFrame(final_indices)
final_index_df.sort_index(inplace=True)
output_filename = "Commodity_price_index/country_export_commodity_price_indices.csv"
final_index_df.to_csv(output_filename)

print(f"\nSuccessfully created and saved country price indices to '{output_filename}'")
print("\n--- Preview of Final Index DataFrame ---")
print(final_index_df.head())
```

# Stats for commodity price indices

```python
import numpy as np
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# --- Load data ---
df_export_price_index = pd.read_csv(
    "Commodity_price_index/country_export_commodity_price_indices.csv",
    index_col="Date",
    parse_dates=True
)

# Restrict sample
df_export_price_index = df_export_price_index.loc["2010-01-01":"2025-12-31"]

# Countries of interest
countries = [
    "Australia", "Brazil", "Canada", "Chile", "Colombia", 
    "Malaysia", "Mexico", "Norway", "Peru", "South Africa"
]
countries = [c for c in countries if c in df_export_price_index.columns]

# --- Prepare Data ---
levels = df_export_price_index[countries]
# Log Levels (for ADF test)
log_levels = np.log(levels)
# Log Returns (for ALL other stats)
log_returns = log_levels.diff().dropna()

# --- Helper Function for ADF Test ---
def get_adf_pvalue(series):
    try:
        # Returns p-value. > 0.05 means Non-Stationary (Unit Root)
        return adfuller(series.dropna(), autolag='AIC')[1]
    except:
        return np.nan

# --- Build the Table ---
# We calculate stats on RETURNS, because that is what matters for risk/shocks
desc_stats = pd.DataFrame(index=countries)

# 1. Volatility (Annualized Standard Deviation)
# We multiply by sqrt(252) to annualize, and 100 to get percentages
desc_stats["Vol. (Ann. %)"] = log_returns.std() * np.sqrt(252) * 100

# 2. Central Tendency (Daily Mean is usually ~0, so Annualized Mean is better)
desc_stats["Mean Ret. (Ann. %)"] = log_returns.mean() * 252 * 100

# 3. Asymmetry (Skewness) - KEY for your "Crash" hypothesis
# Negative skew = frequent small gains, few massive losses (Carry profile)
desc_stats["Skewness"] = log_returns.skew()

# 4. Tail Risk (Kurtosis) - KEY for your "Risk-Off" hypothesis
# High Kurtosis = Extreme events happen often
desc_stats["Kurtosis"] = log_returns.kurt() 

# 5. Persistence (AR1)
desc_stats["AR(1)"] = log_returns.apply(lambda x: x.autocorr(lag=1))

# 6. Econometric Validation (ADF Tests)
# Proof that Levels are I(1) and Returns are I(0)
print("Running ADF tests...")
desc_stats["ADF P-val (Levels)"] = log_levels.apply(get_adf_pvalue)
desc_stats["ADF P-val (Returns)"] = log_returns.apply(get_adf_pvalue)

# --- Formatting ---
desc_stats = desc_stats.round(3)

# --- Save ---
output_file = "Results/export_commodity_descriptives_final.xlsx"
desc_stats.to_excel(output_file)

print("\n--- Final Descriptive Statistics (Based on Log Returns) ---")
print(desc_stats)
```

```python
import pandas as pd
import os

def find_missing_years(filepath, years_to_check):
    """
    Analyzes a commodity price index file to find which countries are
    completely missing data for entire years. (This function remains unchanged).
    """
    if not os.path.exists(filepath):
        print(f"ERROR: File not found at '{filepath}'. Skipping analysis.")
        return None

    print(f"--- Analyzing file: {filepath} ---")
    df_indices = pd.read_csv(filepath, index_col='Date', parse_dates=True)
    all_countries = df_indices.columns
    missing_data_report = {}

    for country in all_countries:
        missing_years_for_country = []
        for year in years_to_check:
            year_str = str(year)
            try:
                yearly_data = df_indices[country].loc[year_str]
                if yearly_data.count() == 0:
                    missing_years_for_country.append(year)
            except KeyError:
                # This handles the case where a year doesn't exist at all in the index
                missing_years_for_country.append(year)
        
        if missing_years_for_country:
            missing_data_report[country] = missing_years_for_country
            
    return missing_data_report

# --- Main Script ---
# Define the expected range of years
YEARS_RANGE = range(2010, 2025)

# Define paths
export_indices_file = "Commodity_price_index/country_export_commodity_price_indices.csv"
import_indices_file = "Commodity_price_index/country_import_commodity_price_indices.csv"
# As per your folder structure, the relative path should be:
country_mapping_file = "Trade/comtrade_country_list.xlsx"

# --- NEW: Load the Country Name -> Code Mapping ---
try:
    country_map_df = pd.read_excel(country_mapping_file)
    # Create the mapping from 'reporterDesc' (name) to 'reporterCode' (code)
    # This is the reverse of what we did in the calculation script
    name_to_code_map = country_map_df.drop_duplicates('reporterDesc').set_index('reporterDesc')['reporterCode'].to_dict()
    print("Successfully loaded country name to reporter code mapping.")
except FileNotFoundError:
    print(f"FATAL ERROR: Country mapping file not found at '{country_mapping_file}'. Reporter codes will not be shown.")
    name_to_code_map = {} # Use an empty dictionary to prevent crashing

print("\n" + "="*70)

# --- Run analysis for the EXPORT file ---
export_missing_report = find_missing_years(export_indices_file, YEARS_RANGE)

if export_missing_report is not None:
    if not export_missing_report:
        print("\n>>> SUCCESS: No completely missing years found for any country in the export file.")
    else:
        print("\n>>> RESULTS: Found missing years in the EXPORT file for the following countries:")
        # --- UPDATED: Print loop with code lookup ---
        for country, years in export_missing_report.items():
            # Look up the code using the country name. Use .get() for safety.
            reporter_code = name_to_code_map.get(country, "N/A") 
            print(f"- Code: {reporter_code:<5} | Country: {country:<30} | Missing Years: {years}")

print("\n" + "="*70 + "\n")

# --- Run analysis for the IMPORT file ---
import_missing_report = find_missing_years(import_indices_file, YEARS_RANGE)

if import_missing_report is not None:
    if not import_missing_report:
        print("\n>>> SUCCESS: No completely missing years found for any country in the import file.")
    else:
        print("\n>>> RESULTS: Found missing years in the IMPORT file for the following countries:")
        # --- UPDATED: Print loop with code lookup ---
        for country, years in import_missing_report.items():
            reporter_code = name_to_code_map.get(country, "N/A")
            print(f"- Code: {reporter_code:<5} | Country: {country:<30} | Missing Years: {years}")

print("\n--- Diagnostic Check Complete ---")
```

```python
import pandas as pd
import os

def renormalize_indices(input_filepath, output_filepath, base_date_str):
    """
    Loads a country index file, re-normalizes all indices to a base of 100
    on a specified date, and saves the result to a new file.

    Args:
        input_filepath (str): Path to the source CSV file.
        output_filepath (str): Path to save the new re-normalized CSV file.
        base_date_str (str): The date to use as the base (e.g., '2016-01-05').
    """
    print(f"--- Processing File: {input_filepath} ---")
    
    # --- 1. Load the Data ---
    try:
        df_indices = pd.read_csv(input_filepath, index_col='Date', parse_dates=True)
    except FileNotFoundError:
        print(f"ERROR: Input file not found. Please check the path: '{input_filepath}'")
        return

    # --- 2. Perform Re-Normalization ---
    try:
        # Get the row of index values on the specified base date
        base_values = df_indices.loc[base_date_str]
        
        # Check if any base values are zero or NaN, which would cause errors
        if (base_values == 0).any() or base_values.isnull().any():
            print(f"WARNING: Some indices have a value of 0 or NaN on the base date {base_date_str}.")
            print("         These columns cannot be re-normalized and will contain NaN values.")

        # Divide the entire DataFrame by the base values, then multiply by 100
        # This scales every country's index so that its value on the base date becomes 100
        renormalized_df = (df_indices / base_values) * 100
        
        print(f"Successfully re-normalized all indices to a base of 100 on {base_date_str}.")

    except KeyError:
        print(f"FATAL ERROR: The base date '{base_date_str}' was not found in the index.")
        print("             Please choose a date that exists in your data. Aborting.")
        return
    except Exception as e:
        print(f"An unexpected error occurred during normalization: {e}")
        return

    # --- 3. Save the New File ---
    renormalized_df.to_csv(output_filepath)
    print(f"Successfully saved re-normalized data to: '{os.path.abspath(output_filepath)}'")
    
    # --- 4. Show a Preview ---
    print("\n--- Preview of Re-Normalized Data ---")
    print(renormalized_df.loc[base_date_str].head().to_string()) # Show the base date values
    print("...")


# --- Main script execution ---

# Define the base date for normalization
BASE_DATE = '2016-01-05'

# Define the input and output file paths
export_input_file = "Commodity_price_index/country_export_commodity_price_indices.csv"
export_output_file = "Commodity_price_index/country_export_commodity_price_indices_normalized.csv"

import_input_file = "Commodity_price_index/country_import_commodity_price_indices.csv"
import_output_file = "Commodity_price_index/country_import_commodity_price_indices_normalized.csv"

# --- Run the process for both export and import files ---
renormalize_indices(export_input_file, export_output_file, BASE_DATE)

print("\n" + "="*60 + "\n")

renormalize_indices(import_input_file, import_output_file, BASE_DATE)

print("\n--- Re-Normalization Complete ---")
```
