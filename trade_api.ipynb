# Download Export data

```python
import os
import time
import pandas as pd
import comtradeapicall
from dotenv import load_dotenv
import math

# --- 1. SETUP ---
# (Setup is the same as before)
load_dotenv()
subscription_key = os.getenv("COMTRADE_API_KEY")

if not subscription_key:
    raise ValueError("API key not found. Make sure it's in your .env file as COMTRADE_API_KEY.")

try:
    commodity_hs_list = pd.read_csv("Trade/commodity_hs_list.csv")["hs"].astype(str).tolist()
    countries_list = pd.read_csv("Trade/country_list.csv")["reporterCode"].astype(str).tolist()
except FileNotFoundError as e:
    raise FileNotFoundError(f"A required CSV file is missing: {e}")

download_period = "2022"

# --- 2. MAIN LOOP WITH RESUMABILITY ---

summary_results = []
output_dir = os.path.join("Trade", "country_data")
os.makedirs(output_dir, exist_ok=True) # Ensure the output directory exists

print(f"Starting/Resuming data download for {len(countries_list)} countries...")

# Loop over each reporter code
for reporter_code in countries_list:
    
    # --- THIS IS THE NEW RESUME LOGIC ---
    filename = f"{reporter_code}_export_data_{download_period}.csv"
    output_path = os.path.join(output_dir, filename)

    if os.path.exists(output_path):
        print(f"\n===== File for Reporter: {reporter_code} already exists. Skipping download. =====")
        
        # Optional: Load the existing data to include it in the final summary
        try:
            existing_df = pd.read_csv(output_path)
            # Perform the calculation on the existing data
            total_exports_row = existing_df[existing_df['cmdCode'] == 'TOTAL']
            if not total_exports_row.empty:
                total_export_value = pd.to_numeric(total_exports_row['primaryValue'], errors='coerce').iloc[0]
                commodity_exports_df = existing_df[existing_df['cmdCode'] != 'TOTAL']
                commodity_export_value = pd.to_numeric(commodity_exports_df['primaryValue'], errors='coerce').sum()
                commodity_share = (commodity_export_value / total_export_value) * 100 if total_export_value > 0 else 0
                summary_results.append({"reporterCode": reporter_code, "commodity_share_percent": commodity_share, "total_export_value": total_export_value})
            else:
                 summary_results.append({"reporterCode": reporter_code, "commodity_share_percent": "No Total", "total_export_value": "No Total"})
        except Exception as e:
            print(f"Could not process existing file for {reporter_code}: {e}")
            summary_results.append({"reporterCode": reporter_code, "commodity_share_percent": "Error", "total_export_value": "Error"})
        
        continue # Move to the next country in the list
    # --- END OF NEW RESUME LOGIC ---

    # If the file doesn't exist, proceed with the download
    print(f"\n=================================================")
    print(f"===== Processing Reporter: {reporter_code} =====")
    print(f"=================================================")
    
    all_dataframes = []
    chunk_size = 150
    # ... (The rest of the download and processing logic is the same as before) ...
    total_chunks = math.ceil(len(commodity_hs_list) / chunk_size)

    for i in range(0, len(commodity_hs_list), chunk_size):
        chunk = commodity_hs_list[i:i + chunk_size]
        cmd_code_string = ",".join(chunk)
        current_chunk_number = (i // chunk_size) + 1
        print(f"--- Requesting Chunk {current_chunk_number}/{total_chunks} for reporter {reporter_code} ---")

        selection_criteria = {
            "typeCode": "C", "freqCode": "A", "clCode": "HS",
            "period": download_period, "reporterCode": reporter_code, "cmdCode": cmd_code_string,
            "flowCode": "X", "partnerCode": "0", "partner2Code": None,
            "customsCode": None, "motCode": None,
        }

        try:
            df_chunk = comtradeapicall.getFinalData(
                subscription_key=subscription_key, **selection_criteria, maxRecords=250000
            )
            if df_chunk is not None and not df_chunk.empty:
                print(f"Downloaded {len(df_chunk)} records.")
                all_dataframes.append(df_chunk)
            else:
                print("No data returned for this chunk.")
            time.sleep(1)
        except Exception as e:
            print(f"An error occurred on chunk {current_chunk_number} for reporter {reporter_code}: {e}")
            continue

    if all_dataframes:
        final_df = pd.concat(all_dataframes, ignore_index=True)
        final_df.to_csv(output_path, index=False) # Save using the path defined earlier
        print(f"Final data for {reporter_code} saved to: {output_path}")

        # Process and add to summary
        final_df['primaryValue'] = pd.to_numeric(final_df['primaryValue'], errors='coerce')
        total_exports_row = final_df[final_df['cmdCode'] == 'TOTAL']
        if not total_exports_row.empty:
            total_export_value = total_exports_row['primaryValue'].iloc[0]
            commodity_exports_df = final_df[final_df['cmdCode'] != 'TOTAL']
            commodity_export_value = commodity_exports_df['primaryValue'].sum()
            commodity_share = (commodity_export_value / total_export_value) * 100 if total_export_value > 0 else 0
            summary_results.append({"reporterCode": reporter_code, "commodity_share_percent": commodity_share, "total_export_value": total_export_value})
        else:
            summary_results.append({"reporterCode": reporter_code, "commodity_share_percent": "No Total", "total_export_value": "No Total"})
    else:
        print(f"\nDownload finished for reporter {reporter_code}, but no data was collected.")
        summary_results.append({"reporterCode": reporter_code, "commodity_share_percent": "No Data", "total_export_value": "No Data"})

# --- 4. FINAL SUMMARY ---

print("\n\n===== ALL COUNTRIES PROCESSED =====")
if summary_results:
    summary_df = pd.DataFrame(summary_results)
    print("Summary of Commodity Export Shares:")
    print(summary_df)
    summary_df.to_csv(f"Trade/commodity_export_share_summary_{download_period}.csv", index=False)
    print(f"\nFull summary saved to Trade/commodity_export_share_summary_{download_period}.csv")
```

```python
### With HS Codes and details 

import os
import pandas as pd

# --- 1. SETUP ---
years_to_process = range(2010, 2025)
flow = "export" 

base_dir = "Trade"
data_dir = os.path.join(base_dir, "country_data")
country_mapping_file = os.path.join(base_dir, "comtrade_country_list.xlsx")

# --- EMU DEFINITIONS ---
# Dictionary defining the year each country joined the EMU.
emu_countries_by_year = {
    'Austria': 2010, 'Belgium': 2010, 'Cyprus': 2010, 'Finland': 2010, 
    'France': 2010, 'Germany': 2010, 'Greece': 2010, 'Ireland': 2010, 
    'Italy': 2010, 'Luxembourg': 2010, 'Malta': 2010, 'Netherlands': 2010, 
    'Portugal': 2010, 'Slovakia': 2010, 'Slovenia': 2010, 'Spain': 2010,
    'Estonia': 2011, 'Latvia': 2014, 'Lithuania': 2015
}

# --- HS CODE DEFINITIONS ---
# (I've placed your lists here for organization)
aluminum_hs = ["260600", "760110", "760120"]
banana_hs = ["080310", "080390", "080410", "080420", "080430", "080440", "080450", "081050", "081060", "200941", "200949", "200820"]
barley_hs = ["100310", "100390", "121010", "121020", "130213"]
beef_hs = ["010221", "010229", "010231", "010239", "010290", "020110", "020120", "020130", "020210", "020220", "020230", "020610", 
           "020621", "020622", "020629", "021020", "051110", "150210", "150290"]
cereal_hs = ["100830", "100840", "100850", "100860", "100890", "110290", "110319", "110320", "110429", "110430", "110510", "110520", 
             "110610", "110620", "110630", "110710", "110720", "121300"]
citrus_hs = ["080510", "080520", "080540", "080550", "080590", "080610", "080620", "080711", "080719", "080720", "081400", "200911", 
             "200912", "200919", "200921", "200929", "200931", "200939", "200830"]
coal_hs = ["270111", "270112", "270119", "270120", "270210", "270220", "270300"]
cobalt_hs = ["260500", "810520"]
cocoa_hs = ["180100", "180200", "180310", "180320", "180400", "180500"]
coffee_hs = ["090111", "090112", "090121", "090122", "090190", "210111", "210112", "210130", "121120", "240110", "240120", "240130", 
             "121294"]
copper_hs = ["260300", "740100", "740200", "740311", "740312", "740313", "740319", "740321", "740322", "740329"]
corn_hs = ["100510", "100590", "151521", "151529", "230690", "230210", "110220", "110313", "110423", "110812"]
cotton_hs = ["520100", "520210", "520291", "520299", "520300", "520411", "520419", "520511", "520512", "520513", "520514", "520515", 
            "520521", "520522", "520523", "520524", "520526", "520527", "520528", "520531", "520532", "520533", "520534", "520535", 
            "520541", "520542", "520543", "520544", "520546", "520547", "520548", "520611", "520612", "520613", "520614", "520615", 
            "520621", "520622", "520623", "520624", "520625", "520631", "520632", "520633", "520634", "520635", "520641", "520642", 
            "520643", "520644", "520645", "120721", "120729", "230610", "530110", "530121", "530129", "530130", "530210", "530290", 
            "530310", "530390"]
crude_oil_hs = ["270900", "271410", "271490"]
shrimp_hs = ["030611", "030612", "030614", "030615", "030616", "030617", "030619", "030621", "030622", "030624", "030625", "030626", 
            "030627", "030629", "030711", "030719", "030721", "030729", "030731", "030739", "030741", "030749", "030751", "030759", 
            "030760", "030771", "030779", "030781", "030789", "030791", "030799", "160510", "160521", "160529", "160530", "160540", 
            "160551", "160552", "160553", "160554", "160555", "160556", "160557", "160558", "160559", "160561", "160562", "160563", 
            "160569", "030811", "030819", "030821", "030829", "030830", "030890", "150410", "150420", "150430", "121221", "121229"]
dairy_hs = ["040110", "040120", "040140", "040150", "040210", "040221", "040229", "040291", "040299", "040310", "040390"]
fertilizers_hs = ["310100", "310510", "310520", "310551", "310559", "310560", "310590"]
fish_hs = ["030191", "030192", "030193", "030194", "030195", "030199", "030211", "030213", "030214", "030219", "030221", "030222", 
           "030223", "030224", "030229", "030231", "030232", "030233", "030234", "030235", "030236", "030239", "030241", "030242", 
           "030243", "030244", "030245", "030246", "030247", "030251", "030252", "030253", "030254", "030255", "030256", "030259", 
           "030271", "030272", "030273", "030274", "030279", "030281", "030282", "030283", "030284", "030285", "030289", "030290", 
           "030311", "030312", "030313", "030314", "030319", "030323", "030324", "030325", "030326", "030329", "030331", "030332", 
           "030333", "030334", "030339", "030341", "030342", "030343", "030344", "030345", "030346", "030349", "030351", "030353", 
           "030354", "030355", "030356", "030357", "030363", "030364", "030365", "030366", "030367", "030368", "030369", "030381", 
           "030382", "030383", "030384", "030389", "030390", "030431", "030432", "030433", "030439", "030441", "030442", "030443", 
           "030444", "030445", "030446", "030449", "030451", "030452", "030453", "030454", "030455", "030459", "030461", "030462", 
           "030463", "030469", "030471", "030472", "030473", "030474", "030475", "030479", "030481", "030482", "030483", "030484", 
           "030485", "030486", "030487", "030489", "030491", "030492", "030493", "030494", "030495", "030499", "030510", "030520", 
           "030531", "030532", "030539", "030541", "030542", "030543", "030544", "030549", "030551", "030559", "030561", "030562", 
           "030563", "030564", "030569", "030571", "030572", "030579", "160300", "160411", "160412", "160413", "160414", "160415", 
           "160416", "160417", "160419", "160420", "160431", "160432"]
fishmeal_hs = ["230120"]
food_hs = ["130211", "200990"]
gold_hs = ["710811", "710812", "710813"]
nuts_hs = ["080111", "080112", "080119", "080121", "080122", "080131", "080132", "080211", "080212", "080221", "080222", "080231", 
           "080232", "080241", "080242", "080251", "080252", "080261", "080262", "080270", "080280", "080290", "120230", "120241", 
           "120242", "230500", "200811", "200819"]
hardwood_hs = ["440341", "440349", "440391", "440392", "440399"]
hides_hs = ["410120", "410150", "410190", "410210", "410221", "410229", "410320", "410330", "410390"]
iron_ore_hs = ["260111", "260112", "720110", "720120", "720150", "720310", "720390", "720610", "720690"]
lamb_hs = ["010410", "010420", "020410", "020421", "020422", "020423", "020430", "020441", "020442", "020443", "020450", "020680", "020690"]
lead_hs = ["260700", "780110", "780191", "780199"]
legumes_hs = ["071310", "071320", "071331", "071332", "071333", "071334", "071335", "071339", "071340", "071350", "071360", "071390", 
              "070810", "070820", "070890", "071021", "071022", "071029", "230250", "121292"]
meat_hs = ["010614", "010633", "020500", "020810", "020830", "020840", "020850", "020860", "020890", "021091", "021092", "021093",
            "021099", "040721", "040729", "040790", "040811", "040819", "040891", "040899", "041000", "050400", "150600", "160100",
            "160210", "160220", "160231", "160232", "160239", "160241", "160242", "160249", "160250", "160290", "040711", "040719",
            "010130", "010190", "010613", "010649", "050210", "050290", "050510", "050590", "050610", "050690", "050800", "051000",
            "230110", "350211", "350219", "350220", "151610"]
# molybdenum_hs = ["260200", "261000", "261100", "261220", "261310", "261390", "261400", "261510", "261590", "261710", "261790", "810194",
#                 "810294", "810320", "810411", "810419", "810720", "810820", "810920", "811010", "811100", "811212", "811221", "811251", "811292"] # 25 hs codes
##### Breaking up molybdenum    One HS-code is removed - description too generic "ores and concentrates, other than antimony"
chromium_hs = ["261000", "811221"]
tungsten_hs = ["261100", "810194"]
thorium_hs = ["261220"]
molybdenum_hs = ["261310", "261390", "810294"]
titanium_hs = ["261400", "810820"]
zirconium_hs = ["261510", "810920"]
niobium_hs = ["261590", "810320"]
antimony_hs = ["261710", "811010"]
tantalum_hs = ["810320"]
magnesium_hs = ["810411", "810419"]
cadmium_hs = ["810720"]
manganese_hs = ["811100"]
beryllium_hs = ["811212"]
thallium_hs = ["811251"]
gallium_hs = ["811292"]


natural_gas_hs = ["271111", "271121"]
nickel_hs = ["260400", "750110", "750120", "750210", "750220"]
nitrogen_fertilizers_hs = ["310210", "310221", "310229", "310230", "310240", "310250", "310260", "310280", "310290"]
apple_hs = ["080810", "080830", "080840", "080910", "080921", "080929", "080930", "080940", "081010", "081020", "081030", "081040", 
            "081070", "081090", "081110", "081120", "081190", "081210", "081290", "081310", "081320", "081330", "081340", "081350", 
            "200961", "200969", "200971", "200979", "200981", "200840", "200850", "200860", "200870", "200880", "200893"]
oats_hs = ["230240", "100410", "100490", "110412", "110422"]
olive_oil_hs = ["150910", "150990", "151000", "071120", "200570", "070992"]
palladium_hs = ["711021", "711029", "711031", "711039"]
palm_oil_hs = ["151110", "151190", "230660", "120300", "120710", "151311", "151319", "151321", "151329", "230650", "530500", "200891"]
phosphate_fertilizers_hs = ["310310", "310390", "310530", "310540"]
platinum_hs = ["711011", "711019"]
pork_hs = ["010310", "010391", "010392", "020311", "020312", "020319", "020321", "020322", "020329", "020630", "020641", "020649", 
           "020910", "020990", "021011", "021012", "021019", "150110", "150120", "150190", "150300"]
potassium_fertilizers_hs = ["310420", "310430", "310490"]
poultry_hs = ["010511", "010512", "010513", "010514", "010515", "010594", "010599", "020711", "020712", "020713", "020714", "020724", 
              "020725", "020726", "020727", "020741", "020742", "020743", "020744", "020745", "020751", "020752", "020753", "020754", 
              "020755", "020760"]
precious_metals_hs = ["261690"]
propane_hs = ["271112", "271113", "271119", "271129"]
rapeseed_oil_hs = ["120510", "120590", "151411", "151419", "151491", "151499", "230641", "230649", "120400", "120740", "120750", "120760", 
                   "120791", "120799", "151511", "151519", "151550", "151590", "230620", "121140"]
rice_hs = ["100610", "100620", "100630", "100640", "230310"]
rubber_hs = ["400110", "400121", "400122", "400129", "400130", "400300", "400400", "400911", "400912", "400921", "400922", "400931", 
             "400932", "400941", "400942"]
sawn_hardwood_hs = ["440110", "440122", "440721", "440722", "440725", "440726", "440727", "440728", "440729", "440791", "440792", 
                    "440793", "440794", "440795", "440799", "440831", "440839", "440929", "440131", "440139", "450110", "450190", 
                    "450200"]
silver_hs = ["261610", "710610", "710691", "710692"]
softwood_hs = ["440121", "440310", "440320", "440410", "440420", "440710", "440810", "440890", "440910", "440921", "140110", "140120"]
sorghum_hs = ["100710", "100790"]
# soybean_hs = ["230400", "150710", "150790", "150810", "150890", "151530", "120110", "120190", "120730", "120810"] # Seperate into just soybean and oil
soybean_hs = ["230400", "120110", "120190", "120730", "120810"] 
soybean_oil_hs = ["150710", "150790", "150810", "150890", "151530"] 
# 
sugar_hs = ["040900", "170112", "170113", "170114", "170191", "170199", "170211", "170219", "170220", "170230", "170240", "170250", 
            "170260", "170290", "170310", "170390", "230320", "010641", "090411", "090412", "090421", "090422", "090510", "090520", 
            "090611", "090619", "090620", "090710", "090720", "090811", "090812", "090821", "090822", "090831", "090832", "090921", 
            "090922", "090931", "090932", "090961", "090962", "091011", "091012", "091020", "091030", "091091", "091099", "120910", 
            "121291", "130120", "130190", "130212", "121293", "130219", "130220", "250100"]
sunflower_seed_oil_hs = ["120600", "151211", "151219", "151221", "151229", "230630", "120770"]
tea_hs = ["090230", "090240", "210120", "090210", "090220", "090300", "121130"]
tin_hs = ["260900", "800110", "800120"]
uranium_hs = ["261210", "284410", "284420"]
vegetable_oil_hs = ["120890", "121190", "151620", "151710"]
vegetables_hs = ["070110", "070190", "070200", "070310", "070320", "070390", "070410", "070420", "070490", "070511", "070519", 
                 "070521", "070529", "070610", "070690", "070700", "070920", "070930", "070940", "070951", "070959", "070960", 
                 "070970", "070991", "070993", "070999", "071010", "071030", "071040", "071080", "071090", "071140", "071151", 
                 "071159", "071190", "071220", "071231", "071232", "071233", "071239", "071290", "071410", "071420", "071430", 
                 "071440", "071450", "071490", "140420", "140490", "200210", "200290", "200410", "200490", "200510", "200520", 
                 "200540", "200551", "200559", "200560", "200580", "200591", "200599", "230800", "200950", "120921", "120922", 
                 "120923", "120924", "120925", "120929", "120930", "120991", "120999", "121299", "121410", "121490", "140190", 
                 "200110", "200310", "200390", "200989", "110813", "110814", "110819", "110820", "200190", "130231", "130232", "130239"]
wheat_hs = ["100111", "100119", "100191", "100199", "110100", "110900", "230230", "100210", "100290", "100810", "100821", "100829", 
            "110311", "110419", "110811"]
wool_hs = ["510111", "510119", "510121", "510129", "510130", "510211", "510219", "510220", "510310", "510320", "510330", "510400", 
           "510510", "510521", "510529", "510531", "510539", "510540", "510610", "510620", "510710", "510720", "510810", "510820", 
           "150500", "500100", "500200", "500300", "500400", "500500"]
zinc_hs = ["260800", "790111", "790112", "790120"]

# A dictionary to hold all HS categories for easy iteration
hs_categories = {
    "aluminum": aluminum_hs, "banana": banana_hs, "barley": barley_hs, "beef": beef_hs,
    "cereal": cereal_hs, "citrus": citrus_hs, "chromium": chromium_hs, "coal": coal_hs, "cobalt": cobalt_hs,
    "cocoa": cocoa_hs, "coffee": coffee_hs, "copper": copper_hs, "corn": corn_hs,
    "cotton": cotton_hs, "crude_oil": crude_oil_hs, "shrimp": shrimp_hs, "dairy": dairy_hs,
    "fertilizers": fertilizers_hs, "fish": fish_hs, "fishmeal": fishmeal_hs,
    "food": food_hs, "gold": gold_hs, "nuts": nuts_hs, "hardwood": hardwood_hs,
    "hides": hides_hs, "iron_ore": iron_ore_hs, "lamb": lamb_hs, "lead": lead_hs,
    "legumes": legumes_hs, "meat": meat_hs, "molybdenum": molybdenum_hs,
    "natural_gas": natural_gas_hs, "nickel": nickel_hs,
    "nitrogen_fertilizers": nitrogen_fertilizers_hs, "apple": apple_hs, "oats": oats_hs,
    "olive_oil": olive_oil_hs, "palladium": palladium_hs, "palm_oil": palm_oil_hs,
    "phosphate_fertilizers": phosphate_fertilizers_hs, "platinum": platinum_hs,
    "pork": pork_hs, "potassium_fertilizers": potassium_fertilizers_hs,
    "poultry": poultry_hs, "precious_metals": precious_metals_hs, "propane": propane_hs,
    "rapeseed_oil": rapeseed_oil_hs, "rice": rice_hs, "rubber": rubber_hs,
    "sawn_hardwood": sawn_hardwood_hs, "silver": silver_hs, "softwood": softwood_hs,
    "sorghum": sorghum_hs, "soybean": soybean_hs, "soybean_oil": soybean_oil_hs, "sugar": sugar_hs,
    "sunflower_seed_oil": sunflower_seed_oil_hs, "thorium": thorium_hs, "tea": tea_hs, "tin": tin_hs, "tungsten": tungsten_hs,
    "uranium": uranium_hs, "vegetable_oil": vegetable_oil_hs,
    "vegetables": vegetables_hs, "wheat": wheat_hs, "wool": wool_hs, "zinc": zinc_hs,
    "titanium": titanium_hs, "zirconium": zirconium_hs, "niobium": niobium_hs, "antimony": antimony_hs, 
    "tantalum": tantalum_hs, "magnesium": magnesium_hs, "cadmium": cadmium_hs, "manganese": manganese_hs,
    "beryllium": beryllium_hs, "thallium": thallium_hs, "gallium": gallium_hs
}

# --- 2. LOAD STATIC MAPPING FILE ---
try:
    country_map_df = pd.read_excel(country_mapping_file)
    # Map from code to name
    code_to_name_map = country_map_df[['reporterCode', 'reporterDesc']].drop_duplicates().set_index('reporterCode')['reporterDesc'].to_dict()
    # Map from name to code (for EMU lookup)
    name_to_code_map = {name: code for code, name in code_to_name_map.items()}
    print("Successfully loaded country mapping files.")
except FileNotFoundError:
    raise FileNotFoundError(f"FATAL ERROR: Country mapping file not found at '{country_mapping_file}'")

# Get the reporter codes for our EMU countries
emu_reporter_codes = {name: name_to_code_map.get(name) for name in emu_countries_by_year.keys()}
print("\nEMU Countries mapped to Reporter Codes:")
for name, code in emu_reporter_codes.items():
    if code is None:
        print(f" - Warning: '{name}' not found in country mapping file.")
print("Mapping complete.")

# --- 3. MAIN PROCESSING LOOP ---
all_years_summary_data = []

for year in years_to_process:
    print(f"\n{'='*20} PROCESSING YEAR: {year} {'='*20}")

    # Initialize a dictionary to aggregate EMU data for the current year
    emu_yearly_aggregator = {}

    total_exports_file = os.path.join(data_dir, f"total_{flow}_all_countries_{year}.xlsx")
    try:
        total_exports_df = pd.read_excel(total_exports_file)
        total_exports_map = total_exports_df.set_index('reporterCode')['primaryValue'].to_dict()
        print(f"Successfully loaded total exports data for {year}.")
    except FileNotFoundError:
        print(f"Warning: Total exports file for {year} not found at '{total_exports_file}'. Skipping this year.")
        continue

    # Determine which countries are in the EMU for the current processing year
    active_emu_codes_for_year = {
        emu_reporter_codes[name] for name, start_year in emu_countries_by_year.items()
        if year >= start_year and emu_reporter_codes.get(name) is not None
    }
    
    try:
        all_files_in_dir = os.listdir(data_dir)
        data_files_for_year = [f for f in all_files_in_dir if f.endswith(f"_{flow}_data_{year}.csv")]
        if not data_files_for_year:
            print(f"Info: No data files found for {year}.")
            continue
    except FileNotFoundError:
        print(f"Warning: Data directory '{data_dir}' not found.")
        continue

    print(f"Found {len(data_files_for_year)} country files to process for {year}...")

    for filename in data_files_for_year:
        try:
            reporter_code = int(filename.split('_')[0])
        except (ValueError, IndexError):
            continue
        
        country_name = code_to_name_map.get(reporter_code, f"Unknown ({reporter_code})")
        value_of_total_exports = total_exports_map.get(reporter_code)

        if value_of_total_exports is None or pd.isna(value_of_total_exports) or value_of_total_exports == 0:
            continue
            
        try:
            file_path = os.path.join(data_dir, filename)
            df = pd.read_csv(file_path, low_memory=False)
            df['primaryValue'] = pd.to_numeric(df['primaryValue'], errors='coerce').fillna(0)
            df['cmdCode'] = df['cmdCode'].astype(str)

            commodities_df = df[
                (df['customsCode'] == 'C00') & (df['motCode'] == 0) &
                (df['partner2Code'] == 0) & (df['cmdCode'] != 'TOTAL')
            ]
            
            unique_commodities_df = commodities_df.drop_duplicates(subset=['cmdCode'])
            value_of_commodity_exports = unique_commodities_df['primaryValue'].sum()
            share_percentage = (value_of_commodity_exports / value_of_total_exports) * 100
            
            row_data = {
                "Year": year, "Country code": reporter_code, "Country name": country_name,
                "Value of commodity exports (in USD)": value_of_commodity_exports,
                "Value of total exports (in USD)": value_of_total_exports,
                "Share of exports from commodities (percentage)": share_percentage
            }
            
            for category_name, hs_codes in hs_categories.items():
                category_df = commodities_df[commodities_df['cmdCode'].isin(hs_codes)]
                category_value = category_df['primaryValue'].sum()
                
                # --- ADJUSTMENT: Calculate share against COMMODITY exports ---
                if value_of_commodity_exports > 0:
                    category_share_percentage = (category_value / value_of_commodity_exports) * 100
                else:
                    category_share_percentage = 0.0 # Avoid division by zero
                
                row_data[f"Value_{category_name}_USD"] = category_value
                row_data[f"Share_{category_name}_percent_of_commodity"] = category_share_percentage
            
            all_years_summary_data.append(row_data)

            # --- ADDITION: Aggregate data if the country is an active EMU member ---
            if reporter_code in active_emu_codes_for_year:
                for key, value in row_data.items():
                    # We only want to sum the numeric value columns
                    if isinstance(value, (int, float)):
                        emu_yearly_aggregator[key] = emu_yearly_aggregator.get(key, 0) + value

        except Exception as e:
            print(f"An error occurred while processing '{filename}': {e}")
            continue

    # --- ADDITION: After processing all countries for a year, create the EMU aggregate row ---
    if emu_yearly_aggregator and emu_yearly_aggregator.get("Value of total exports (in USD)", 0) > 0:
        print(f"--- Creating EMU Aggregate for {year} ---")
        emu_total_exports = emu_yearly_aggregator["Value of total exports (in USD)"]
        emu_commodity_exports = emu_yearly_aggregator["Value of commodity exports (in USD)"]

        emu_row_data = {
            "Year": year, "Country code": 999, "Country name": "EMU",
            "Value of total exports (in USD)": emu_total_exports,
            "Value of commodity exports (in USD)": emu_commodity_exports
        }
        
        # Calculate overall commodity share for EMU
        emu_row_data["Share of exports from commodities (percentage)"] = (emu_commodity_exports / emu_total_exports) * 100

        # Calculate shares for each HS category for EMU
        for category_name in hs_categories.keys():
            value_key = f"Value_{category_name}_USD"
            share_key = f"Share_{category_name}_percent_of_commodity"
            
            category_value = emu_yearly_aggregator.get(value_key, 0)
            emu_row_data[value_key] = category_value
            
            if emu_commodity_exports > 0:
                emu_row_data[share_key] = (category_value / emu_commodity_exports) * 100
            else:
                emu_row_data[share_key] = 0.0
                
        all_years_summary_data.append(emu_row_data)

# --- 4. CREATE AND SAVE THE FINAL DATAFRAME ---
if all_years_summary_data:
    final_summary_df = pd.DataFrame(all_years_summary_data)
    
    main_cols = [
        "Year", "Country code", "Country name", "Value of total exports (in USD)",
        "Value of commodity exports (in USD)", "Share of exports from commodities (percentage)"
    ]
    category_cols = sorted([col for col in final_summary_df.columns if col not in main_cols])
    final_summary_df = final_summary_df[main_cols + category_cols]

    final_summary_df = final_summary_df.sort_values(
        by=["Year", "Share of exports from commodities (percentage)"], 
        ascending=[True, False]
    ).reset_index(drop=True)

    print("\n--- Final Summary with EMU (First 5 rows, few columns) ---")
    print(final_summary_df.iloc[:5, :10].to_string())
    
    # Check for EMU row in the last year processed to confirm aggregation
    last_year = max(years_to_process)
    print(f"\n--- EMU Data for {last_year} ---")
    print(final_summary_df[(final_summary_df['Year'] == last_year) & (final_summary_df['Country name'] == 'EMU')].iloc[:, :10].to_string())

    output_filename = os.path.join(base_dir, f"final_commodity_{flow}_summary_all_years_WITH_EMU.csv")
    final_summary_df.to_csv(output_filename, index=False)
    
    print(f"\nSuccessfully created and saved the detailed summary with EMU to '{output_filename}'")
else:
    print("\nNo data was processed.")
```

```python
import os
import pandas as pd

# --- 1. SETUP ---
year = "2024"
flow = "import"
data_dir = os.path.join("Trade", "country_data")
country_mapping_file = "Trade/comtrade_country_list.xlsx"
# Construct the full path to the total imports file
total_imports_file = os.path.join(data_dir, "total_import_all_countries.xlsx") 

# --- 2. LOAD MAPPING FILES ---

# Load the country code to country name mapping
try:
    country_map_df = pd.read_excel(country_mapping_file)
    country_map = country_map_df[['reporterCode', 'reporterDesc']].drop_duplicates().set_index('reporterCode')['reporterDesc'].to_dict()
    print("Successfully loaded country mapping file.")
except FileNotFoundError:
    raise FileNotFoundError(f"Error: Country mapping file not found at '{country_mapping_file}'")

# Load the definitive total imports values
try:
    total_imports_df = pd.read_excel(total_imports_file)
    total_imports_map = total_imports_df.set_index('reporterCode')['primaryValue'].to_dict()
    print(f"Successfully loaded total imports data from '{total_imports_file}'.")
except FileNotFoundError:
    raise FileNotFoundError(f"Error: Total imports file not found at '{total_imports_file}'")

# --- 3. PROCESS EACH COUNTRY'S DATA FILE ---
try:
    all_files = os.listdir(data_dir)
    data_files = [f for f in all_files if f.endswith(f"_{flow}_data_{year}.csv")]
    if not data_files:
        raise FileNotFoundError(f"No '{flow}' data files for the year {year} found in '{data_dir}'")
except FileNotFoundError:
    raise FileNotFoundError(f"Error: The data directory was not found: '{data_dir}'")

summary_data = []
print(f"\nProcessing {len(data_files)} country files with new logic...")

for filename in data_files:
    try:
        reporter_code = int(filename.split('_')[0])
    except (ValueError, IndexError):
        print(f"Warning: Could not parse reporter code from '{filename}'. Skipping.")
        continue
    
    country_name = country_map.get(reporter_code, f"Unknown ({reporter_code})")
    print(f"\n--- Processing {country_name} ---")
    
    # --- STEP 1: GET THE TOTAL imports (DENOMINATOR) FROM THE DEDICATED FILE ---
    value_of_total_imports = total_imports_map.get(reporter_code)

    if value_of_total_imports is None:
        print(f"Warning: Could not find total imports value for {country_name} (Code: {reporter_code}). Skipping.")
        continue
    
    if value_of_total_imports == 0:
        print(f"Warning: Total imports for {country_name} is zero. Skipping.")
        continue
        
    print(f"Found total imports from source file: ${value_of_total_imports:,.0f}")

    # --- STEP 2: CALCULATE COMMODITY imports (NUMERATOR) FROM THE COUNTRY FILE ---
    try:
        file_path = os.path.join(data_dir, filename)
        df = pd.read_csv(file_path, low_memory=False)
        df['primaryValue'] = pd.to_numeric(df['primaryValue'], errors='coerce').fillna(0)

        # Filter for the correct type of rows for commodities
        commodities_df = df[
            (df['customsCode'] == 'C00') & 
            (df['motCode'] == 0) &
            (df['partner2Code'] == 0) &
            (df['cmdCode'] != 'TOTAL')
        ]
        
        # --- THIS IS THE FIX ---
        # Remove duplicate rows for the same commodity, keeping only the first instance
        unique_commodities_df = commodities_df.drop_duplicates(subset=['cmdCode'])
        
        # Now, sum the values from the de-duplicated DataFrame
        value_of_commodity_imports = unique_commodities_df['primaryValue'].sum()

        # --- STEP 3: CALCULATE THE FINAL SHARE ---
        share_percentage = (value_of_commodity_imports / value_of_total_imports) * 100
        
        print(f"Calculated unique commodity value: ${value_of_commodity_imports:,.0f} | Share: {share_percentage:.2f}%")
        
        summary_data.append({
            "Country code": reporter_code,
            "Country name": country_name,
            "Value of commodity imports (in USD)": value_of_commodity_imports,
            "Value of total imports (in USD)": value_of_total_imports,
            "Share of imports from commodities (percentage)": share_percentage
        })

    except Exception as e:
        print(f"An error occurred while processing commodity data in '{filename}': {e}")
        continue

# --- 4. CREATE AND SAVE THE FINAL DATAFRAME ---
if summary_data:
    final_summary_df = pd.DataFrame(summary_data)
    final_summary_df = final_summary_df.sort_values(by="Share of imports from commodities (percentage)", ascending=False).reset_index(drop=True)

    print("\n--- Final Summary DataFrame ---")
    print(final_summary_df.to_string())
    
    output_filename = f"Trade/final_commodity_{flow}_summary_{year}_FINAL.csv"
    final_summary_df.to_csv(output_filename, index=False)
    
    print(f"\nSuccessfully created and saved the final summary to '{output_filename}'")
else:
    print("\nNo data was processed.")
```

```python
import os
import pandas as pd

# --- 1. SETUP: CONFIGURE YOUR CHECK HERE ---

# The country code you want to investigate
country_code_to_check = 36 # 8, 24, 28, 31, 32, 36, 40, 51, 52, 56, 60, 68, 70, 76, 84, 96

# The HS commodity codes you want to sum up (enter as strings)
hs_codes_to_check = ["270111", "270112", "270119", "270120", "270210", "270220", "270300"]

# Set the flow to 'export' or 'import'
flow = "export" 

# --- You don't need to change anything below this line ---

# --- 2. FILE AND PATH SETUP ---
year = "2024"
data_dir = os.path.join("Trade", "country_data")
country_mapping_file = "Trade/comtrade_country_list.xlsx"

# Construct the filename for the specific country's data
filename = f"{country_code_to_check}_{flow}_data_{year}.csv"
file_path = os.path.join(data_dir, filename)

# --- 3. LOAD COUNTRY MAPPING ---
try:
    country_map_df = pd.read_excel(country_mapping_file)
    country_map = country_map_df[['reporterCode', 'reporterDesc']].drop_duplicates().set_index('reporterCode')['reporterDesc'].to_dict()
    print("Successfully loaded country mapping file.")
except FileNotFoundError:
    print(f"Error: Country mapping file not found at '{country_mapping_file}'. Exiting.")
    # We exit here because we can't get the name without the file.
    exit()

# --- 4. LOAD AND PROCESS THE DATA ---

# Get the country name from the map we just loaded
country_name = country_map.get(country_code_to_check, f"Unknown Code")

print(f"\n--- Checking HS Codes for {country_name} (Code: {country_code_to_check}) ---")
print(f"Loading data from: '{file_path}'")

try:
    # Load the specific country's data file
    # Ensure cmdCode is read as a string to preserve leading zeros
    df = pd.read_csv(file_path, low_memory=False, dtype={'cmdCode': str})
    df['primaryValue'] = pd.to_numeric(df['primaryValue'], errors='coerce').fillna(0)
    
    # --- 5. APPLY THE EXACT SAME LOGIC AS THE MAIN SCRIPT ---
    
    # Filter for the correct type of rows (World total, standard customs, all transport)
    filtered_df = df[
        (df['customsCode'] == 'C00') & 
        (df['motCode'] == 0) &
        (df['partner2Code'] == 0) &
        (df['cmdCode'] != 'TOTAL')
    ].copy()
    
    # Remove duplicate commodity rows, keeping only the first valid entry
    unique_commodities_df = filtered_df.drop_duplicates(subset=['cmdCode'])
    
    # --- 6. FIND AND SUM THE SPECIFIED HS CODES ---
    
    total_value = 0
    found_codes = []

    print("\n--- Individual HS Code Values ---")
    for code in hs_codes_to_check:
        # Find the row for the specific HS code in our cleaned data
        code_row = unique_commodities_df[unique_commodities_df['cmdCode'] == code]
        
        if not code_row.empty:
            value = code_row['primaryValue'].iloc[0]
            total_value += value
            found_codes.append({'code': code, 'value': value})
            print(f"- HS Code {code}: ${value:,.2f}")
        else:
            print(f"- HS Code {code}: Not Found in the filtered data.")
            
    # --- 7. PRINT THE FINAL SUM ---
    print("-----------------------------------")
    print(f"Sum of found HS codes: ${total_value:,.2f}")
    print("-----------------------------------")

except FileNotFoundError:
    print(f"\nError: Data file not found at '{file_path}'.")
    print("Please ensure the country code, year, and flow are correct.")
except Exception as e:
    print(f"\nAn error occurred: {e}")
```
