```python
import os
import pandas as pd
import numpy as np

# Define the input directory
input_directory = 'Currency'

# --- Step 1: Group files by the original currency pair name ---
file_groups = {}
try:
    files = os.listdir(input_directory)
    for filename in files:
        if filename.endswith('.csv'):
            currency_pair = filename.split('_')[0]
            file_path = os.path.join(input_directory, filename)
            file_groups.setdefault(currency_pair, []).append(file_path)
except FileNotFoundError:
    print(f"Error: The directory '{input_directory}' was not found.")
    file_groups = {}

# --- Step 2: Process each group and standardize the name ---
# This dictionary will group the final Series by their standardized name
# e.g., {'AEDUSD': [series_from_AEDUSD, series_from_USDAED]}
standardized_series_groups = {}

if not file_groups:
    print("No CSV files found in the directory.")
else:
    for currency_pair, file_list in file_groups.items():
        print(f"Processing group: {currency_pair}")
        try:
            list_of_dfs = [pd.read_csv(f, parse_dates=['timestamp']) for f in file_list]
            combined_df = pd.concat(list_of_dfs)

            if combined_df.empty:
                print(f"  - Skipping {currency_pair} as its files are empty.")
                continue

            combined_df.set_index('timestamp', inplace=True)
            combined_df.sort_index(inplace=True)
            combined_df = combined_df[~combined_df.index.duplicated(keep='last')]

            daily_close = combined_df['close'].resample('D').last().ffill()

            # Determine the standardized name and apply inversion if needed
            if currency_pair.startswith('USD'):
                standardized_name = currency_pair[3:] + 'USD'
                print(f"  - Inverting to create standardized series: {standardized_name}")
                daily_close = 1 / daily_close.replace(0, np.nan)
                daily_close.name = standardized_name
            else:
                standardized_name = currency_pair
                daily_close.name = standardized_name
            
            # Add the processed series to the correct standardized group
            standardized_series_groups.setdefault(standardized_name, []).append(daily_close)

        except Exception as e:
            print(f"  - Could not process group {currency_pair}. Error: {e}")

# --- Step 3: Resolve conflicts by picking the series with the earliest data ---
final_series_to_combine = []
print("\n--- Resolving Duplicates ---")

for standardized_name, series_list in standardized_series_groups.items():
    if len(series_list) == 1:
        # No conflict, just add the series
        final_series_to_combine.append(series_list[0])
    else:
        # CONFLICT: We have multiple series for the same name (e.g., from AEDUSD and USDAED)
        print(f"Conflict found for {standardized_name}. Found {len(series_list)} versions. Selecting the one with the earliest data.")
        
        best_series = None
        earliest_date = pd.Timestamp.max # Initialize with the latest possible date

        for series in series_list:
            # Find the first valid (non-empty) date in the series
            first_valid_date = series.first_valid_index()
            if first_valid_date is not None and first_valid_date < earliest_date:
                earliest_date = first_valid_date
                best_series = series
        
        if best_series is not None:
            print(f"  - Selected version starting on {earliest_date.date()}")
            final_series_to_combine.append(best_series)

# --- Step 4: Final Combination and Save ---
if final_series_to_combine:
    print("\nCombining all unique, standardized currency data into a single file...")
    final_df = pd.concat(final_series_to_combine, axis=1)
    final_df.sort_index(axis=1, inplace=True)

    output_filename = 'all_currencies_standardized_daily_close.csv'
    final_df.to_csv(output_filename)

    full_path = os.path.abspath(output_filename)
    print(f"\nProcessing complete. All unique daily data saved to:\n{full_path}")
else:
    print("\nNo data was processed. The output file was not created.")
```

```python
import pandas as pd
import time

# List of desired currency country codes
country_codes = [
    "AR", "AU", "AT", "BE", "BR", "CA", "CL", "CN", "CO", "CZ",
    "DK", "FI", "FR", "DE", "GR", "HK", "HU", "IN", "ID", "IE",
    "IT", "JP", "KR", "MY", "MX", "NL", "NZ", "NO", "PE", "PH",
    "PL", "PT", "RU", "SA", "SG", "ZA", "ES", "SE", "CH", "TH",
    "TR", "GB", "US"
]

# --- 1. Construct the full series keys ---
# We will request Daily, Nominal, Broad EERs. You can change 'N' to 'R' for Real EER.
series_keys = [f"D.N.B.{code}" for code in country_codes]

# --- 2. Break the keys into smaller chunks for multiple API calls ---
def get_chunks(lst, n):
    """Yield successive n-sized chunks from lst."""
    for i in range(0, len(lst), n):
        yield lst[i:i + n]

# Let's request 10 currencies at a time to keep the URL length reasonable
chunk_size = 10
key_chunks = list(get_chunks(series_keys, chunk_size))

all_data_frames = []
base_url = "https://stats.bis.org/api/v2/data/dataflow/BIS/WS_EER/1.0/"
start_period = "2010-01-01"
end_period = "2025-12-31"

print(f"Preparing to download data for {len(series_keys)} series in {len(key_chunks)} chunks...")
print("-" * 30)

# --- 3. Loop through the chunks and download data for each one ---
for i, chunk in enumerate(key_chunks):
    try:
        # Join the keys in the current chunk with a '+'
        key_string = "+".join(chunk)
        
        # Construct the full URL for this chunk
        url = f"{base_url}{key_string}?startPeriod={start_period}&endPeriod={end_period}&format=csv"
        
        print(f"Downloading chunk {i+1}/{len(key_chunks)}...")
        # print(f"URL: {url}") # Uncomment this line if you want to see each URL
        
        # Download the data for the current chunk
        df_chunk = pd.read_csv(url)
        all_data_frames.append(df_chunk)
        
        print(f"-> Success for chunk {i+1}.")
        
        # It's polite to wait a moment between requests to not overload the server
        time.sleep(1) 

    except Exception as e:
        print(f"-> FAILED for chunk {i+1}. Error: {e}")
        print(f"   Problematic keys: {chunk}")

# --- 4. Combine all downloaded chunks and save ---
if all_data_frames:
    # Concatenate all the individual DataFrames into one large DataFrame
    final_df = pd.concat(all_data_frames, ignore_index=True)
    
    print("\nAll chunks downloaded and combined successfully!")
    print("First 5 rows of the final data:")
    print(final_df.head())
    
    try:
        output_file = "bis_eers_daily_currency_data.csv"
        final_df.to_csv(output_file, index=False)
        print(f"\nData successfully saved to '{output_file}'")
    except Exception as e:
        print(f"\nAn error occurred while saving the file: {e}")
else:
    print("\nNo data was downloaded. Please check the error messages above.")
```
